```python
# ::Task1::
# Relationship Between Alarm_Bit and Fault_Status

plt.figure(figsize=(20,5))
plt.plot(df[' Alarm_Bit'],df[' Fault_status'], color = 'r', label='Fault Status')

plt.xticks(rotation=60)
plt.legend()

plt.title('Realtionship Between Alarm_Bit and Fault_Status ')
plt.xlabel('Fault_Status')
plt.ylabel('Alarm_Bit')

plt.show()
```

```python
# ::Task2::
# When blocked status is true , good count is constant , running status is false

plt.figure(figsize=(20,3))
plt.plot(df[' Good_Count'],df[' Blocked_Status'], color = 'r', label='Blocked_Status')
plt.plot(df[' Good_Count'],df[' Running_Status'], color = 'y', label='Running_Status')

plt.xticks(rotation=60)
plt.legend()

plt.title('Realtionship Between Alarm_Bit and Fault_Status ')
plt.xlabel('Fault_Status')
plt.ylabel('Alarm_Bit')

plt.show()
```

```python
# ::Task3::
# Stopped Status and running status are complimentory

plt.figure(figsize=(20,3))
plt.plot(df[' Stopped_Status'],df[' Running_Status'], color = 'y')
plt.xticks(rotation=60)

plt.title('Realtionship Between Stopped Status and Running Status ')
plt.xlabel('Stopped Status')
plt.ylabel('Running Status')

plt.show()
```

```python
# ::Task4::
# Stopped Status and Stop Button should be same

plt.figure(figsize=(20,3))
plt.plot(df[' Stopped_Status'],df[' Stop Button'], color = 'r')
plt.xticks(rotation=60)

plt.title('Realtionship Between Stopped Status and Stop Button ')
plt.xlabel('Stopped Status')
plt.ylabel('Stop Button')

plt.show()
```

```python
# WaitingCode = non-zero, running status = false , stopped status = true, good count = not increasing

group5 = df.groupby(' WaitingCode')
group5.first()
```

```python
# When WaitingCode is non-zero
gp5_1 = group5[[' Running_Status',' Good_Count',' Stopped_Status']].get_group(1)
gp5_1.plot()
```

```python
# When WaitingCode is zero
gp5_2 = group5[[' Running_Status',' Good_Count',' Stopped_Status']].get_group(0)
gp5_2.plot()
```
