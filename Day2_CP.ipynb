```python
import pandas as py
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = py.read_csv('timestamp_fault_17_01_2020.csv')
# df
```

```python
# ::Task6::
# Showing the fault status for a particular timestamp for 17 Jan 2020

plt.figure(figsize=(25,7))
plt.bar(df['time'],df[' Fault_status'],width=0.5, color = 'y', label='Fault_status')

plt.xticks(rotation=90)

# plt.rcParams.update({'font.size': 22})
plt.title('Realtionship Between Timestamp and Fault Status')
plt.xlabel('Time')
plt.ylabel('Fault Status')

frequency = 15
x = df['time']
y = df[' Fault_status']
y1 = np.arange(y.shape[0])
# plt.plot(x,y)
plt.xticks(y1[::frequency], x[::frequency])

plt.show()
```

```python
# ::Task7::
# Showing the fault status for a particular timestamp for few datas

plt.figure(figsize=(25,7))
plt.bar(df['time'].iloc[0:50],df[' Fault_status'].iloc[0:50],width=0.5, color = 'y', label='Fault_status')

plt.xticks(rotation=90)

# plt.rcParams.update({'font.size': 22})
plt.title('Realtionship Between Timestamp and Fault Status')
plt.xlabel('Time')
plt.ylabel('Fault Status')

frequency = 15
x = df['time'].iloc[0:50]
y = df[' Fault_status'].iloc[0:50]
y1 = np.arange(y.shape[0])
plt.plot(x,y)
# plt.xticks(y1[::frequency], x[::frequency])

plt.show()
```
